$y = 0.0015×x_1 + 0.99×x_2$
$y = W^Tx$
0.99和0.0015为回归系数
2. 平方误差
$\sum_i{y_i-W^Tx_i}$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_file = './ex0.txt'
data = []
label = []
with open(data_file) as f:
    lines = f.readlines()
    for line in lines:
        temp = line.strip().split('\t')
        data.append(temp[:2])
        label.append(temp[-1])
```

```python
print(data[:10])
print(label[:10])
data_mat = np.array(data, dtype=np.float)
label_mat = np.array(label, dtype=np.float)
print(data_mat[:10])
print(label_mat[:10])
```

```python
data_num = len(data)
plt.figure(figsize=(10, 10))
plt.scatter(data_mat[:,1], label_mat)
plt.show()
```

```python
def standRegres(X, Y):
    xTx = np.dot(X.T, X)
    if np.linalg.det(xTx) == 0.0:
        print("This matrix is singular, cannot do inverse.")
        return
    W = np.dot(xTx.I, np.dot(X.T, np.mat(Y).T))
    return W
```

```python
W = standRegres(data_mat, label_mat)
```

```python
W
```

```python
y = data_mat*W
print(y)
```

```python
plt.scatter(data_mat[:,1], label_mat)
plt.plot(data_mat[:,1], y, c='black')
plt.show()
```
